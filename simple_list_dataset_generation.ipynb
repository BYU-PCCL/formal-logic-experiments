```python
def prepend_article(word):
    if not word:
        raise ValueError("Word cannot be empty")

    first_letter = word[0].lower()

    if first_letter in "aeiou":
        return "an " + word
    else:
        return "a " + word
```

```python
def generate_training_example(greater_than, relationship, less_than):
    return f"{prepend_article(greater_than.lower())} {relationship} {prepend_article(less_than.lower())}".capitalize()
```

```python
generator: np.random.Generator
```

```python
def generate_training_examples(n, relationship):
    global generator
    # Randomly sample training examples without replacement
    training_objects = generator.choice(OBJECTS, n)
    # Shuffle the training examples
    generator.shuffle(training_objects)
    training_examples = []
    test_examples = {}
    for i in range(n - 1):
        greater_than, less_than = training_objects[i], training_objects[i + 1]
        training_examples.append(
            generate_training_example(greater_than, relationship, less_than)
        )
    # Generate all possible pairs for test examples
    for distance in range(1, n):
        test_examples[distance] = []
        for i in range(n - distance):
            greater_than, less_than = (
                training_objects[i],
                training_objects[i + distance],
            )
            # Check if the pair exists in training examples
            if not any(
                [
                    greater_than == ex[0] and less_than == ex[2]
                    for ex in training_examples
                ]
            ):
                test_examples[distance].append(
                    generate_training_example(greater_than, relationship, less_than)
                )
    return training_examples, test_examples
```

```python
generator = np.random.default_rng(12345)
generate_training_examples(20, "is more blivergent than")
```
